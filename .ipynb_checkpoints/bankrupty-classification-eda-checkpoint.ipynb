# EDA of Bankruptcy Data and Building SVM, Logit, and Naive Bayes Models

The goal is to explore the new data set and build three models (using an 80/20 train/test split) to predict bankruptcy of a firm.

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns

from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, RocCurveDisplay, auc, PrecisionRecallDisplay
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

## Ingest

```python
data = pd.read_csv('https://raw.githubusercontent.com/jhancuch/bankruptcy-classification-eda/main/data/data.csv')
```

## EDA

### Dependent and Independent Variable Distributions

#### Dependent Variable

```python
print('Dependent Variable Summary Statistics:\n{0}\n'.format(data['Bankrupt?'].describe()))
print('Dependent Variable Median: {0}'.format(data['Bankrupt?'].median()))
print('Dependent Variable Mean: {0}\n'.format(data['Bankrupt?'].mean()))

plt.hist(data['Bankrupt?'], bins = 2)
plt.title("Bar Chart of Bankrupt?")
plt.show()
```

#### Independent Variables

```python
data.hist(bins = 100, figsize = (30, 20))
plt.show()
```

### Relationship between Bankruptcy? and Independent Variables

#### Output

```python
for i in range(1, len(data), 4):
    end = i + 4
    temp = data.iloc[:, i:end].merge(data.iloc[:, 0], how='left', left_index=True, right_index=True)
    if len(temp.columns) == 1:
        break
    sns.pairplot(data=temp, hue="Bankrupt?")
    plt.show()
```

## Modeling

### Split Train/Validation and Scale

```python
features = list(data.iloc[:, 1:].columns)
```

```python
X_train_temp, X_validation_temp, y_train, y_validation = train_test_split(data.iloc[:, 1:], data.iloc[:, 0], test_size=0.2, random_state=123)
```

```python
standard_scaler = StandardScaler()
```

```python
X_train = standard_scaler.fit_transform(X_train_temp)

X_validation = standard_scaler.fit_transform(X_validation_temp)
```

### SVM

#### Hyperparameter tuning

```python
param_grid = {'C': [0.1, 0.5, 1, 5, 10],
              'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}
```

```python
grid = GridSearchCV(SVC(), param_grid, scoring = 'roc_auc', verbose = 1, n_jobs = 2)       
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

#### Fitted Model and Analysis

commentary - even with grid search, it leaves a bit to be desired since non bankrupt companies comprise of x% of our dataset so just guessing not bankrupt would result in %

```python
svm = make_pipeline(SVC(C=5, kernel='linear')).fit(X_train, y_train.ravel())
```

```python
svm_predictions = svm.predict(X_validation)
```

```python
print(classification_report(y_validation, svm_predictions))
print('Accuracy: {0}\n'.format(accuracy_score(y_validation, svm_predictions)))

CM = confusion_matrix(y_validation, svm_predictions)

TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]

print('True Positive Rate {0}\n'.format(TP/(TP+FN)))
print('True Negative Rate {0}\n'.format(TN/(TN+FP)))
```

ROC/AUC -- Commentary

```python
fpr, tpr, thresholds = roc_curve(y_validation, svm_predictions)
roc_auc = auc(fpr, tpr)
```

```python
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                  estimator_name='SVM')
```

```python
display.plot()
plt.show()
```

Precision / Recall Plot -- commentary

```python
display = PrecisionRecallDisplay.from_predictions(y_validation, svm_predictions, name="SVC")
_ = display.ax_.set_title("2-class Precision-Recall curve")
```

### Logistic

#### Hyperparamter tuning

```python
param_grid = {'penalty': ['l2', 'none'],
              'C': [.001, .01, .1, 1, 5, 10],
              'max_iter': [4000]}
```

```python
grid = GridSearchCV(LogisticRegression(random_state = 123), param_grid, scoring = 'roc_auc', verbose = 1, n_jobs = 2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

#### Fitted Model and Analysis

commentary - even with grid search, it leaves a bit to be desired since non bankrupt companies comprise of x% of our dataset so just guessing not bankrupt would result in %

```python
logit = make_pipeline(LogisticRegression(C=0.01, max_iter=4000, random_state=123)).fit(X_train, y_train.ravel())
```

```python
logit_predictions = logit.predict(X_validation)
```

```python
print(classification_report(y_validation, logit_predictions))
print('Accuracy: {0}\n'.format(accuracy_score(y_validation, logit_predictions)))

CM = confusion_matrix(y_validation, logit_predictions)

TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]

print('True Positive Rate {0}\n'.format(TP/(TP+FN)))
print('True Negative Rate {0}\n'.format(TN/(TN+FP)))
```

ROC commentary

```python
fpr, tpr, thresholds = roc_curve(y_validation, logit_predictions)
roc_auc = auc(fpr, tpr)
```

```python
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                  estimator_name='Logit')
```

```python
display.plot()
plt.show()
```

Precision/Recall Plot Commentary

```python
display = PrecisionRecallDisplay.from_predictions(y_validation, logit_predictions, name="SVC")
_ = display.ax_.set_title("2-class Precision-Recall curve")
```
